##Define a function to match the columns, Hubspot and Sage

```python
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

def check_word_in_column(sheet, source_col, target_col, output_col):
    """
    Checks if a word in the source column exists in the target column and writes 1 for matched,
    0 for not matched, and 0.5 for partial matches.

    Args:
    - sheet: The worksheet object.
    - source_col: The index of the source column (1-based).
    - target_col: The index of the target column (1-based).
    - output_col: The index of the column to store the match results (1-based).
    """

    # Define fill colors for matches, non-matches, and partial matches
    match_fill = PatternFill(start_color="C6EFCE", end_color="C6EFCE", fill_type="solid")  # Green for matched
    partial_fill = PatternFill(start_color="FFA500", end_color="FFA500", fill_type="solid")  # Orange for partial match
    no_match_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")  # Yellow for unmatched

    # Get all values from the target column
    target_values = {str(sheet.cell(row=row, column=target_col).value) for row in range(2, sheet.max_row + 1)}

    for row in range(2, sheet.max_row + 1):  # Assuming row 1 contains headers
        source_value = sheet.cell(row=row, column=source_col).value  # Value from the source column
        source_value_str = str(source_value) if source_value is not None else ""

        # Check if the source value is in the target values
        if source_value_str in target_values:
            match_result = 1  # Matched
            sheet.cell(row=row, column=output_col).fill = match_fill  # Apply green fill for matched
        elif any(source_value_str in str(target_value) for target_value in target_values):
            match_result = 0.5  # Partial match
            sheet.cell(row=row, column=output_col).fill = partial_fill  # Apply orange fill for partial matches
        else:
            match_result = 0  # Not matched
            sheet.cell(row=row, column=output_col).fill = no_match_fill  # Apply yellow fill for unmatched

        # Write the result to the output column
        sheet.cell(row=row, column=output_col).value = match_result
```

##Load the workbook and select the worksheet

```python
wb = load_workbook('/content/hubspot&sagedata.xlsx')
ws = wb["Sheet1"]

# Define column indices (1-based)
source_col = 1   # words in Column A (hubspot)
target_col = 2   # target column to search in Column B(sage)
output_col = 3   # Column C(match results)

# Check if words from source column are in target column
check_word_in_column(ws, source_col, target_col, output_col)

# Save the file
wb.save("word_match_hubspot_sage.xlsx")
```

```python
import pandas as pd

def match_sheets_with_condition(sheet1_path, sheet2_path, key_column, condition_column, condition_value):
    # Load both sheets into DataFrames
    sheet1_df = pd.read_excel(sheet1_path)
    sheet2_df = pd.read_excel(sheet2_path)

    # Merge on the key column to find matches
    matched_df = pd.merge(sheet1_df, sheet2_df, on='Record ID', how='left')

    # Apply the condition and create a new column with the result
    matched_df['Match_Result'] = matched_df[condition_column].apply(
        lambda x: "1" if x == condition_value else "0"
    )

    return matched_df
```
